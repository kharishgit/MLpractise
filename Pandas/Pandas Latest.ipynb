# Creating DataFrame from JSON

```python
iw = pd.DataFrame({
    "city":['Mumbai','Delhi','Kolkatta'],
    "Temparature":[32,40,38],
    "Humidity":[80,60,78]
})
iw
```

```python
uw = pd.DataFrame({
    "city":['Newyork','Chicago','Orlando'],
    "Temparature":[2,15,36],
    "Humidity":[8,62,50]
})
uw
```

# Concatinating

```python
df = pd.concat([iw,uw])
```

```python
df
```

```python
df = pd.concat([iw,uw],ignore_index= True)
df
```

# Adding New Keys

```python
df = pd.concat([iw,uw],keys = ["India","USA"])
df
```

```python
df.loc['India']
```

# Appending Columns

```python
tmp =  pd.DataFrame({
    "city":['Newyork','Chicago','Orlando'],
    "Temparature":[2,15,36]
    
})
tmp
```

```python
winspeed =  pd.DataFrame({
    "city":['Newyork','Chicago','Orlando'],
    "Winspeed":[10,7,14]
    
})
winspeed
```

```python
df = pd.concat([tmp,winspeed],axis = 1)
df
```

# Pandas Series


```python
import pandas as pd
import numpy as np
```

```python
# help(pd.Series)
```

```python
myindex = ["USA","India","Canada"]
```

```python
mydata = [1776,1947,1821]
```

```python
myser = pd.Series(data = mydata)
type(myser)
```

```python
myser = pd.Series(data = mydata, index = myindex)
myser
```

```python
myser['India']
```

```python
# Also can be written in Order
myser = pd.Series(mydata,myindex)
myser
```

```python
# Creating from dictionary

ages = {'sam':6,'ramu':8,'raju':4}
ser = pd.Series(ages)
ser
```

```python
sales_q1 = {
    "USA": 150000,
    "Germany": 120000,
    "Japan": 130000,
    "India": 110000
}
q1 = pd.Series(sales_q1)
q1
```

```python
sales_q2 = {
    "USA": 170000,
    "Spain": 125000,
    "Japan": 135000,
    "India": 115000
}
q2 = pd.Series(sales_q2)
q2
```

```python
# Math Operations
q1/1000
```

```python
q2 * 15
```

```python
#  Adding
q1 + q2
```

```python
q1.add(q2,fill_value =0)
```
