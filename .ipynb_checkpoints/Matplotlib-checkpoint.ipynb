```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.arange(0,9)
```

```python
y= 3*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlim(0,7)
plt.ylim(0,20)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title');
plt.savefig('matplot1.png')
```

```python
a = np.linspace(0,10,11)
```

```python
a
```

```python
b= a**4
```

```python
b
```

```python
fig = plt.figure(dpi=100,figsize=(10,5))
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.3,0.5,0.5])
axes2.plot(x,y)
axes1.plot(a,b);
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
axes
```

```python
axes.shape
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(b,a)
axes[1][0].plot(y,x)
axes[1][1].plot(a,b)
plt.tight_layout()
```
