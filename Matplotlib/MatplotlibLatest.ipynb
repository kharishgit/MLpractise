```python
days = [1,2,3,4,5,6]
avg_tmp =[21,30,25,41,5,37]
max_tmp = [26,35,32,46,10,41]
min_tmp = [17,24,19,36,0,17]
```

```python
plt.title('Weather')
plt.ylabel('Days')
plt.plot(days,avg_tmp,label = 'Avg')
plt.plot(days,max_tmp,label = 'Max')
plt.plot(days,min_tmp,label = 'Min')
plt.legend(loc = "best")
plt.grid()
```

```python
company = ['GOGL','FB','AMZN','MCRSFT']
revenue = [120,65,89,56]
profit = [19,10,30,40]
xpos = np.arange(len(company))
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width = 0.4, label = 'Revenue')
plt.bar(xpos+0.2,profit,width = 0.4,label = 'Profit')
plt.legend()
```

```python
company = ['GOGL','FB','AMZN','MCRSFT']
revenue = [120,65,89,56]
profit = [19,10,30,40]
plt.yticks(xpos,company)
xpos = np.arange(len(company))
plt.xticks(xpos,company)
plt.barh(xpos-0.2,revenue, label = 'Revenue')
plt.barh(xpos+0.2,profit,label = 'Profit')
plt.legend()
```
